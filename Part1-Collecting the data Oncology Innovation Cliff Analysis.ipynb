# Phase 1:  Collecting the data


In 2026, Pharmaceutical companies are facing a "innovation cliff." It takes over 10 years and $2.5 billion to bring a single drug to market. The most critical juncture is Phase II Clinical Trials, where the drug is tested for efficacy in humans for the first time. Currently, most companies rely on "siloed" data scientists look at lab results while business teams look at market trends-leading to fragmented decision-making.

Source: https://biotech-spain.com/en/articles/c-suite-playbook-pharmaceutical-market-research-trends-2026/



## Objectives


The Phase II failure rate is approximately 70%.

- Scientific Risk: Molecules that worked in a petri dish often fail in the complex human body.

- Economic Risk: Even if a drug works, it might fail commercially if a competitor launches a cheaper alternative or if the manufacturing cost is too high.

The Gap: There is a lack of Unified Data Architecture that combines biological, financial, and competitive data to predict a "Go/No-Go" decision before the money is spent.

## Data Wrangling


We can see below that some of the rows are missing values in our dataset.


```python
print(df_lean.isnull().sum())
```

Observation: The dataset contains no null values across all primary strategic variables.

Analysis: While "zero missing values" may appear anomalous, it is a byproduct of regulatory data integrity. In the pharmaceutical sector, ClinicalTrials.gov enforces strict schema validation; if mandatory fields like the NCT ID, Study Status, or Sponsor are not populated, the record is rejected at the source.

Additionally, our current view represents a pre-filtered slice of the global database. The system likely suppressed incomplete or "draft" entries during the export process to ensure the public has access only to actionable, high-quality data.
